```python
import random
import pandas as pd
from sklearn.cluster import KMeans, AgglomerativeClustering, SpectralClustering, BisectingKMeans, MiniBatchKMeans, SpectralBiclustering, SpectralCoclustering
from sklearn.model_selection import ParameterGrid
from sklearn import metrics
import matplotlib.pyplot as plt

def listToStr(ls):
    output = ""
    for l in ls:
        output += str(l) + ","
    return output[0:len(output)-1]

def loadDF():
    df = pd.read_csv("oliveira.csv")
    df = df[df['malware'] == 1].copy()
    df = df.drop('malware', axis=1)
    return df.reset_index().iloc[:,1:]

def getX(df):
    return df.iloc[:, 1:102-1]

#Load list of API calls
API_LIST = "api_calls.txt"
DELIMITER = "NaN"
API_FILE = open(API_LIST,"r")
APIS = API_FILE.readline().split(',')
APIS.append(DELIMITER) #serves as a label for NaN values for Instance-based datasets
API_FILE.close()

#Random Seed
seed = 1

#Inverse Label Encoding
def inverse_labeller(item):
    global APIS
    return item.map(lambda x: APIS[int(x)])
def inverse_label(df):
    df.iloc[:, 1:101] = df.iloc[:, 1:101].apply(inverse_labeller, axis=1, result_type='reduce')
    return df
```

From: https://medium.com/swlh/k-means-clustering-on-high-dimensional-data-d2151e1a4240

```python
def searchK(parameters, model, X):
    # instantiating ParameterGrid, pass number of clusters as input
    parameter_grid = ParameterGrid({'n_clusters': parameters})
    best_score = -1
    best_grid = -1
    best_clusterer = None
    silhouette_scores = []
    # evaluation based on silhouette_score
    print("Searching best cluster size...")
    for p in parameter_grid:
        model.set_params(**p)    # set current hyper parameter
        model.fit_predict(X)          # fit model on wine dataset, this will find clusters based on parameter p
        ss = metrics.silhouette_score(X, model.labels_)   # calculate silhouette_score
        silhouette_scores += [ss]       # store all the scores
        print('Parameter:', p, 'Score', ss)
        # check p which has the best score
        if ss > best_score:
            best_score = ss
            best_grid = p
            best_clusterer = model
    # plotting silhouette score
    print("Plotting results...")
    plt.bar(range(len(silhouette_scores)), list(silhouette_scores), align='center', color='#722f59', width=0.5)
    plt.xticks(range(len(silhouette_scores)), list(parameters))
    plt.title('Silhouette Score', fontweight='bold')
    plt.xlabel('Number of Clusters')
    plt.show()
    return best_grid, best_clusterer

def clustering(inner_df, name, clusterer, clusters):
    X = getX(inner_df)
    bestCluster, bestClusterer = searchK(clusters, clusterer, X)
    #bestClusterer.fit(X)
    inner_df['cluster'] = bestClusterer.fit_predict(X)
    inner_df = inverse_label(inner_df)
    inner_df.to_csv(f"Clustering/{name}_Clustering_{bestCluster['n_clusters']:.0f}.csv", index=False)
    return bestCluster['n_clusters']

def commonAPICluster(inner_df, name, size):
    inner_df = df
    clusters = inner_df['cluster'].unique()
    clusters.sort()
    columns = inner_df.columns[1:102]
    commonAPI = []
    print("Searching for Common API Patterns per Cluster...")
    print(clusters)
    for cluster in clusters:
        commonC = inner_df[inner_df['cluster']==cluster]['pattern']#.value_counts()
        common = [cluster, commonC.value_counts()[0], round(commonC.value_counts()[0]/commonC.shape[0],4), commonC.iloc[0]]
        commonAPI.append(common)
    commonAPI = pd.DataFrame(commonAPI, columns=['cluster', 'count', 'match ratio', 'pattern'])
    commonAPI.to_csv(f"Clustering/{name}_CommonAPICluster_{size}.csv", index=False)
    return commonAPI

def getSampleHash_Common(inner_df, common_counts, name, size, samplesize):
    hashes = []
    global seed
    random.seed(seed)
    matching_samples = 0
    print(f"Random (Seed @ {seed}) Sampling Hashes subset of  Most Common API Patterns...")
    for pattern in range(common_counts.shape[0]):
        sub_df = inner_df[inner_df['pattern']==common_counts.iloc[pattern,3]]
        subsamples = len(sub_df['hash'].to_list())
        matching_samples += subsamples
        if subsamples < samplesize:
            subsamples = random.sample(sub_df['hash'].to_list(), len(sub_df['hash'].to_list()))
        else:
            subsamples = random.sample(sub_df['hash'].to_list(), samplesize)
        for subsample in subsamples:
            hashes.append([common_counts.iloc[pattern,:]['cluster'], subsample, common_counts.iloc[pattern,:]['pattern']])
    hashes = pd.DataFrame(hashes, columns=['cluster', 'hash', 'pattern'])
    hashes.to_csv(f"Clustering/{name}_SampleHash_Common_{size}.csv", index=True)
    print(f"Commonality Ratio: {(matching_samples/inner_df.shape[0])*100:.4f}%")
    return hashes
    
def injectPatterns(inner_df, inverse_labelled_df):
    patterns = []
    for row in range(inner_df.shape[0]):
        patterns.append(listToStr(inverse_labelled_df.iloc[row,1:101].transpose().to_list()))
    inner_df['pattern'] = patterns
    return inner_df
```

# Load Dataset

```python
df = injectPatterns(loadDF(), inverse_label(loadDF()))
df
```

# Explanation

The process as the following:
1. Find the best cluster size (by means of the [Silhouette Score](https://tushar-joshi-89.medium.com/silhouette-score-a9f7d8d78f29))
2. Search for the most common API pattern for each cluster
3. Sample n hashes per cluster that match the most common API pattern for the same cluster.

**Commonality Ratio**: How many of the matching most common API patterns per cluster are there relative to the dataset size. *(Higher is better)*

```python
reserve_df = df.copy()
```

# K-Means

```python
df = reserve_df.copy()
print(df.shape)

clusters = [20,40,60,80,100] #Let's assume that there are up to 100 clusters that can be derived from the Oliveira dataset. Going beyond 100 is too much
samplesize = 10

#Search for best clusters size and run clustering
bestClusterSize = clustering(df, "KMeans", KMeans(n_init='auto', verbose=0, random_state=seed), clusters)

commonAPI = commonAPICluster(df, "KMeans", bestClusterSize)
display(commonAPI)

commonHashes = getSampleHash_Common(df, commonAPI, "KMeans", bestClusterSize, samplesize)
display(commonHashes)
```

# BisectingKMeans

```python
df = reserve_df.copy()
print(df.shape)

clusters = [20,40,60,80,100] #Let's assume that there are up to 100 clusters that can be derived from the Oliveira dataset. Going beyond 100 is too much
samplesize = 10

#Search for best clusters size and run clustering
bestClusterSize = clustering(df, "BisectingKMeans", BisectingKMeans(init='k-means++', random_state=seed, verbose=0, copy_x=True, algorithm='lloyd', bisecting_strategy='biggest_inertia'), clusters)

commonAPI = commonAPICluster(df, "BisectingKMeans", bestClusterSize)
display(commonAPI)

commonHashes = getSampleHash_Common(df, commonAPI, "BisectingKMeans", bestClusterSize, samplesize)
display(commonHashes)
```

# MiniBatchKMeans

```python
df = reserve_df.copy()
print(df.shape)

clusters = [20,40,60,80,100] #Let's assume that there are up to 100 clusters that can be derived from the Oliveira dataset. Going beyond 100 is too much
samplesize = 10

#Search for best clusters size and run clustering
bestClusterSize = clustering(df, "MiniBatchKMeans", MiniBatchKMeans(init='k-means++', verbose=0, random_state=seed, tol=0.001, max_no_improvement=100, n_init='auto'), clusters)

commonAPI = commonAPICluster(df, "MiniBatchKMeans", bestClusterSize)
display(commonAPI)

commonHashes = getSampleHash_Common(df, commonAPI, "MiniBatchKMeans", bestClusterSize, samplesize)
display(commonHashes)
```
