# Oliveira Merge

Combines a malware type labeled dataset and a non-labelled one (i.e., oliveira.csv)

This also corrects the `malware` column based on the type value. 

```python
import pandas as pd

import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)

oliveira = pd.read_csv('./oliveira.csv')
labels = pd.read_csv('Labelling_DBSCAN_Clustering.csv')
```

```python
print("How many falsely labelled malicious samples?:", labels[labels['Type 1']=='_'].shape[0])
```

```python
label_list = labels['hash'].to_list()
label = []
for i in range(oliveira.shape[0]):
    if oliveira.iloc[i,0] in label_list:
        idx = labels[labels['hash'] == oliveira.iloc[i,0]].index
        t = labels.iloc[idx,3].to_string().split('    ')[1].strip()
        # if t == '_':
        #     oliveira.iloc[i,101] = 0
        label.append(t)
    else:
        label.append('benign')
oliveira['type'] = pd.Series(label)
oliveira
```

```python
oliveira.to_csv('oliveira_labelled.csv', index=False)
```
