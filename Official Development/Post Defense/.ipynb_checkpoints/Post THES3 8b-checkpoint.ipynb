## Load Libraries

```python
import pandas as pd
import lightgbm as lgbm
import catboost as catb

from imblearn import under_sampling as us

from sklearn.model_selection import train_test_split as ts
from sklearn.metrics import classification_report, roc_auc_score, accuracy_score, average_precision_score, roc_auc_score, recall_score, precision_score, f1_score
from sklearn.metrics import ConfusionMatrixDisplay, confusion_matrix, RocCurveDisplay

from joblib import load, dump
import warnings
warnings.filterwarnings("ignore")
```

## Load DF

```python
def load_training(df):
    X, X_t, y, y_t = ts(df[df['type'] == 'trojan'].iloc[:,1:], df[df['type'] == 'trojan']['malware'], train_size=0.07, random_state=1, shuffle=True)
    X['malware'] = y
    df = df[df['type'] != 'trojan']
    df = pd.concat([df, X], ignore_index=True)
    df.reset_index(inplace=True)
    df.drop(columns=['index'], inplace=True)
    return df.copy(deep=True)

def load_test(df):
    #df = df[df['malware'] == 0]
    df = df.reset_index()
    df = df.drop(columns='index')
    return df.copy(deep=True)
```

```python
#LGBM Data

#Training
lgbm_df = load_training(pd.read_csv("../Dataset/TB/LGBM_TB_PreSMOTE.csv", low_memory=False))
display("Training Malware Type Distribution", lgbm_df['type'].value_counts())

#Testing
lgbm_test_df = load_test(pd.read_csv("../Dataset/TB/LGBM_TB_Test.csv", low_memory=False))
```

```python
#CATB Data

#Training
catb_df = load_training(pd.read_csv("../Dataset/TB/CATB_TB_PreSMOTE.csv", low_memory=False))
display("Training Malware Type Distribution", lgbm_df['type'].value_counts())

#Testing
catb_test_df = load_test(pd.read_csv("../Dataset/TB/CATB_TB_Test.csv", low_memory=False))
```

## Train Models

```python
def get_indexes():
    indexes = []
    for i in range(100):
        indexes.append(f"{i}")
    return indexes

lgbm_model = lgbm.LGBMClassifier(random_state=1, n_jobs=0, verbose=-1, categorical_feature=get_indexes())
lgbm_model.fit(X=lgbm_df.iloc[:,1:101], y=lgbm_df['malware'])
dump(lgbm_model, "lightgbm.model")
```

```python
def get_indexes():
    indexes = []
    for i in range(100):
        indexes.append(f"t_{i}")
    return indexes

catb_model = catb.CatBoostClassifier(random_state=1, thread_count=-1, verbose=1, cat_features=get_indexes(), 
                                     nan_mode='Min', one_hot_max_size=256, task_type='GPU')
catb_model.fit(X=catb_df.iloc[:,1:101], y=catb_df['malware'])
catb_model.save_model("catboost.model",format="json")
```

## Test Models

```python
def evaluate(model, df):
    y_pred = model.predict(df.iloc[:,1:101])
    y_actual = df['malware']
    print(classification_report(y_actual, y_pred, digits=4))
    print(roc_auc_score(y_actual, y_pred))
```

## Load Predictions

```python
lgbm_predictions = evaluate(lgbm_model, lgbm_test_df)
```

```python
catb_predictions = evaluate(catb_model, catb_test_df)
```
