```python
# Import Libraries
import pandas as pd

df = pd.read_csv('./MalBD.csv')
df.dropna(inplace=True)
print(df.shape)
df.head()
```

```python
MALBD_APIS = []

for i in range(df.shape[0]):
    MALBD_APIS += df.iloc[i,2:].to_list()
MALBD_APIS = list(set(MALBD_APIS))

MALBD_APIS.sort()
print(len(MALBD_APIS))
print(MALBD_APIS)
```

```python
#Load list of API calls
DELIMITER = "NaN"
API_FILE = open('./Dataset/api_calls.txt',"r")
APIS = API_FILE.readline().split(',')
APIS.append(DELIMITER) #serves as a label for NaN values for Instance-based datasets
API_FILE.close()

APIS.sort()
print(len(APIS))
print(APIS)
```

```python
indices = []
for i in range(df.shape[0]):
    ls = df.iloc[i,2:].to_list()
    if not set(ls) <= set(APIS):
        indices.append(i)
print(len(indices))

df.to_csv('MalBD - RealWorld.csv', index=False)
```
