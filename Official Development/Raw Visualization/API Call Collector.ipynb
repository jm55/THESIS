```python
import pandas as pd

df = pd.read_csv('../Dataset/oliveira_labelled.csv')
df_str = pd.read_csv('../Dataset/oliveira_labelled_str.csv')

display(df)
print("")
display(df_str)
```

```python
cols = df.iloc[:,1:101].columns.to_list()
```

```python
def print_calls(df, df_str):
    out = []
    for c in cols:
        print(c, df_str[c].value_counts().index.to_list()[0], df[c].value_counts().index.to_list()[0], df_str[c].value_counts().to_list()[0])
        out.append(df_str[c].value_counts().index.to_list()[0])
    return pd.Series(out).unique().tolist()
```

```python
print_calls(df[df['malware']==0], df_str[df_str['malware']==0])
```

```python
print_calls(df[df['malware']==1], df_str[df_str['malware']==1])
```
